Testing stationary

```python
#H0: It SI NON STATIONARY
#H1: IT IS STATIONARY
from statsmodels.tsa.stattools import adfuller
test_result=adfuller(df['Paper_Prod'])
def adfuller_test(prd):
  result=adfuller(prd)
  labels=['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
  for value,label in zip(result,labels):
    print(label+':'+str(value))
  if result[1]<=0.05:
    print("Strong evidence against the null hypothesis(Ho), reject the null hypothesis")
  else:
    print("Weak evidence against null hypothesis, time series has a unit root indicating it as non stationary")
adfuller_test(df['Paper_Prod'].dropna())
```

Moving average and Weighted Moving Average

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
sns.set_theme()
df=pd.read_csv('paper_production.csv',parse_dates=True)
df['Period']=pd.to_datetime(df['Period'], format='%b-%y')
df.set_index('Period',inplace=True)
df["Sales"]=df["Paper_Prod"].rolling(window=2,min_periods=1).mean()
weights = [0.1, 0.3, 0.6]
window_size = len(weights)
df["WMA"] = (
    df["Paper_Prod"]
    .rolling(window=window_size)
    .apply(lambda x: np.dot(x, weights), raw=True)
).shift(1)
df["sma5"]=df["Paper_Prod"].rolling(window=5,min_periods=1).mean()
print(df)
sns.lineplot(data=df, x="Period",y="Sales")
df['em0.3']=df["Paper_Prod"].ewm(alpha=0.2,adjust=False).mean()
df['em0.5']=df["Paper_Prod"].ewm(alpha=0.5,adjust=False).mean()

df[["Paper_Prod","WMA",'sma5','em0.3','em0.5']].plot(figsize=(12,8))
```

```python
!pip install pulp # Install the pulp library
from pulp import * # Import all functions from the pulp module
#Sources and Destinations:List
Br=['A','B','C','D']
WH=[1,2,3,4]
```

```python
#Supply from each Branch
s={'A':35,'B':50,'C':80,'D':65}
#Demand from each warehouse
d={1:70,2:30,3:75,4:55}
#Cost for all Branches and Warehouses
C={'A':{1:10,2:7,3:6,4:4},
   'B':{1:8,2:8,3:5,4:7},
   'C':{1:4,2:3,3:6,4:9},
   'D':{1:7,2:5,3:4,4:3}
   }
```

```python
#Setting Problem
prob=LpProblem("Transportation",LpMinimize)
```

```python
routes=[(i,j) for i in Br for j in WH]
#Defining Decision Variables
amt_var=LpVariable.dicts("Amountship",(Br,WH),0)
```

```python
#Defining Objective Function
prob+=lpSum(amt_var[i][j]*C[i][j] for (i,j) in routes)
#Constraints
for j in WH:
  prob+=lpSum(amt_var[i][j] for i in Br) >=d[j]
for i in Br:
  prob+=lpSum(amt_var[i][j] for j in WH) <=s[i]
```

```python
prob.solve()
```

```python
print("Status:", LpStatus[prob.status])
```

```python
for v in prob.variables():
  if v.varValue >0:
    print(v.name,"=",v.varValue)
```

Holt's Linear Trend Model, also known as Holt's Trend-Corrected Exponential Smoothing, is a time series forecasting technique that extends simple exponential smoothing by incorporating a trend component. This makes it suitable for time series data that exhibit a trend but do not show seasonality. It was introduced by Charles Holt in 1957.

Key Components of Holts Linear Trend Model
The model has two main components:

Level (L): The baseline value of the time series, which adjusts over time.
Trend (T): The direction and rate of change in the time series.
The model uses two smoothing equations to update these components iteratively:

Level Equation: Updates the current level based on the observed value and the estimated trend.
Trend Equation: Updates the trend based on the difference between the current and previous levels.

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, Holt
import seaborn as sns
from matplotlib import pyplot as plt
sns.set_theme()
df=pd.read_csv('/content/paper_production.csv',parse_dates=True)
df['Period']=pd.to_datetime(df['Period'], format='%b-%y')
df.set_index('Period',inplace=True)
train=df.iloc[:-int(len(df)*0.2)]
test=df.iloc[-int(len(df)*0.2):]
#Fit simple model amd get Forecasts
model_simple=SimpleExpSmoothing(train['Paper_Prod']).fit(optimized=True)
forecasts_simple=model_simple.forecast(len(test))
#Fit holt linear model
model_holt=Holt(train['Paper_Prod'],damped_trend=True).fit(optimized=True)
forecasts_holt=model_holt.forecast(len(test))
plt.figure(figsize=(12,8))
plt.plot(train.index,train['Paper_Prod'],label='Train',color='red')
plt.plot(test.index,test['Paper_Prod'],label='Test',color='blue')
plt.plot(forecasts_simple.index,forecasts_simple.values,label='Simple',color='yellow')
plt.plot(forecasts_holt.index,forecasts_holt.values,label='Holt',color='green')
plt.title("Overlaying Multiple PLots")
plt.xlabel("Period")
plt.ylabel("Paper Production")
plt.legend()
plt.grid()
plt.show()
```

HOLT WINTERS

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, Holt, ExponentialSmoothing
import seaborn as sns
from matplotlib import pyplot as plt
sns.set_theme()
df=pd.read_csv('/content/paper_production.csv',parse_dates=True)
df['Period']=pd.to_datetime(df['Period'], format='%b-%y')
df.set_index('Period',inplace=True)
train=df.iloc[:-int(len(df)*0.2)]
test=df.iloc[-int(len(df)*0.2):]
#Fit simple model amd get Forecasts
model_simple=SimpleExpSmoothing(train['Paper_Prod']).fit(optimized=True)
forecasts_simple=model_simple.forecast(len(test))
#Fit holt linear model
model_holt=Holt(train['Paper_Prod'],damped_trend=True).fit(optimized=True)
forecasts_holt=model_holt.forecast(len(test))
#Fit holt's winter model
model_holt_winter=ExponentialSmoothing(train['Paper_Prod'],trend='add',seasonal='add',seasonal_periods=12).fit()
forecasts_holt_winter=model_holt_winter.forecast(len(test))
#PLot the forcast
plt.figure(figsize=(12,8))
plt.plot(train.index,train['Paper_Prod'],label='Train',color='red')
plt.plot(test.index,test['Paper_Prod'],label='Test',color='blue')
plt.plot(forecasts_simple.index,forecasts_simple.values,label='Simple',color='yellow')
plt.plot(forecasts_holt.index,forecasts_holt.values,label='Holt',color='green')
plt.plot(forecasts_holt_winter.index,forecasts_holt_winter.values,label='Holt Winter',color='orange')
plt.title("Overlaying Multiple PLots")
plt.xlabel("Period")
plt.ylabel("Paper Production")
plt.legend()
plt.grid()
plt.show()
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1['Age'],df1['Income($)'],color='red')
plt.scatter(df2['Age'],df2['Income($)'],color='green')
plt.scatter(df3['Age'],df3['Income($)'],color='black')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
scaler=MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)']=scaler.transform(df[['Income($)']])
scaler.fit(df[['Age']])
df['Age']=scaler.transform(df[['Age']])
df
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
km.cluster_centers_
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1['Age'],df1['Income($)'],color='red',label='Income1')
plt.scatter(df2['Age'],df2['Income($)'],color='green',label='Income2')
plt.scatter(df3['Age'],df3['Income($)'],color='black',label='Income3')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
k_rng=range(1,10)
sse=[]
for k in k_rng:
  km=KMeans(n_clusters=k)
  km.fit(df[['Age','Income($)']])
  sse.append(km.inertia_)
```
